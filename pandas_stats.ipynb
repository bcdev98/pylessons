```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
# open file
filename = "incomes.csv"

with open(filename, "r") as fobj:
    contents = fobj.read()

print(contents)
```

```python
with open(filename, "r") as fileObj:
    for line in fileObj:
        print(line)
```

```python
x_values = [0, 1, 2, 3, 4, 5, 6, 7]
squares = [0, 1, 4, 9, 16, 25, 36, 49]
plt.plot(x_values, squares)
plt.show()
```

```python
x_values = list(range(1000))
cubes = [x**3 for x in x_values]
plt.scatter(x_values, cubes, s=10)
plt.show()
```

```python
x = np.linspace(0, 10)
line = plt.plot(x, np.sin(x), 'o', linewidth=3)
plt.title("simple sin plot")
plt.show()
```
